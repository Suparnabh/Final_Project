```python
# Dependencies
from stats import mean, median, mode, multi_mode
import matplotlib.pyplot as plt
```

```python
def plot_data(fig_num, data):
    plt.figure(fig_num)
    plt.bar(range(len(data)),data)
```

```python
def plot_mean_med(fig_num, data):
    plt.figure(fig_num)
    plt.hlines(mean(data), xmin=-1, xmax=len(data), color='g', label='mean')
    plt.hlines(median(data), xmin=-1, xmax=len(data), color='r', label='median')
    plt.legend()
    plt.show()
```

```python
def plot_modes(fig_num, data):
    plt.figure(fig_num)
    modes = multi_mode(data)
    for m in modes:
        plt.hlines(m[0], xmin=-1, xmax=len(data), color='b', label='mode')
    plt.legend()
    plt.show()
```

## Measures of Central Tendency

```python
data = [30, 31, 31, 32, 32, 40, 41, 41, 300]
data
```

```python
plot_data(0, data)
```

```python
plot_mean_med(0,data)
```

```python
data = [30, 31, 31, 32, 32, 40, 41, 41, 300, 350, 350, 468]
plot_data(1, data)
```

```python
plot_mean_med(1, data)
```

```python
plot_modes(1, data)
```

```python
# Prices of random electronics at Best Buy
prices = [4, 425, 984, 2932, 49]
print(f"Median Price: {median(prices)}")
print(f"Mean Price: {mean(prices)}")
```
