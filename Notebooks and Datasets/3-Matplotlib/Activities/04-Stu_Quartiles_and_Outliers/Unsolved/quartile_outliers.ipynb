```python
# Dependencies
import matplotlib.pyplot as plt
from stats import median
import numpy as np
import seaborn as sns
```

```python
### Data Points
arr = np.array([2.3, 10.2,11.2, 12.3, 14.5, 14.6, 15.0, 15.1, 19.0, 24.0])
arr
```

```python
# Find the median
np.percentile(arr, 50)
```

```python
median(arr)
```
